```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.simplefilter(action='ignore',category=FutureWarning)
```

```python
dataset = pd.read_csv("Algerian.csv",header =1 )
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
#datacleaning
```

```python
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
df=dataset
df.loc[:122,"region"] = 0
df.loc[122:,"region"] = 1
```

```python
df.info()
```

```python
df[["region"]] = df[["region"]].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#remove the null values
df = df.dropna().reset_index(drop = True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
##fix spaces in cols name
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
#convert obj to int/float
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
##changing the other cols to float datatype
```

```python
objects = [features for features in df.columns if df[features].dtypes=="O"]
```

```python
for i in objects:
  if i!= "Classes": #skipped classes as its catogerical or target
    df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
objects
```

```python
df.describe().T
```

```python
#to save clean
df.to_csv('Algerian_cleaned_Dataset',index=False)
```
